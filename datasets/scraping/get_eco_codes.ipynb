# get_eco_codes.ipynb

## This notebook outlines the process for creating a CSV file which contains the ECO code and opening name (with additional information) for display in opening information.

The first thing I did was go to [https://www.chessgames.com/chessecohelp.html](https://www.chessgames.com/chessecohelp.html) to get a table with all of the ECO codes and description with lines for scraping.
I saved it to a file named 'ECO Listing.html'.

Next, I import BeautifulSoup to extract the relevant information from the page.

```python
from bs4 import BeautifulSoup
with open('ECO Listing.html') as fp:
    soup = BeautifulSoup(fp, 'html.parser')
```

```python
with open('eco_codes_test.csv', 'w') as eco_fp:
    eco_fp.write('ECO,opening\n')
    for tr in soup.find_all('tr')[:]:
        tds = tr.find_all('td')
        eco_fp.write("{},\"{} ({})\"\n".format(tds[0].text,
                                               tds[1].text.split('\n')[0],
                                               tds[1].text.split('\n')[1]))
```

```python
with open('eco_codes_test.csv', 'r') as e_fp:
    for line in e_fp:
        print(line)
```
